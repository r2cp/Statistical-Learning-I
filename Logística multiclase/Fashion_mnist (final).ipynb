## Selección de las clases y obtención de los datos de entrenamiento
Se utilizarán las categorías 0, 1 y 7, correspondientes a las camisetas y tops, pantalones y zapatillas. La clase 7 será remapeada a la etiqueta **2** para poder codificar en one-hot, o bien, utilizar la función de TensorFlow *tf.nn.sparse_softmax_cross_entropy_with_logits*.

A continuación se define el diccionario para hacer referencia a las clases. 

```python
# Función para seleccionar las clases 
def seleccionarClases(train_labels, classes):
    # Obtener el filtro de clases
    f1 = train_labels == classes[0]
    f2 = train_labels == classes[1]
    f3 = train_labels == classes[2]
    f = f1 | f2 | f3
    
    # Filtrar las etiquetas
    YtrainLabels = train_labels[f]
    
    # Renombrar las clases a las etiquetas 0, 1, 2
    YtrainLabels[YtrainLabels == classes[0]] = 0
    YtrainLabels[YtrainLabels == classes[1]] = 1
    YtrainLabels[YtrainLabels == classes[2]] = 2
    
    return f, YtrainLabels

# Función para obtener las imágenes correspondientes a las etiquetas seleccionadas
# con la función anterior. Para esto se utiliza el parámetro de filtro
def obtenerImagenes(train_images, filtro):
    XtrainImages = np.reshape(train_images, (train_images.shape[0], 28*28))
    return XtrainImages[filtro]
```

#### Generar las etiquetas y los datos de entrenamiento

```python
# Obtener las etiquetas 
filtroLabels, YtrainLabels = seleccionarClases(train_labels, [0, 1, 7])
# Obtener las imágenes asociadas a las etiquetas
XtrainImages = obtenerImagenes(train_images, filtroLabels)
```

```python
# Conjunto de entrenamiento: (# observaciones, # features)
N, K = XtrainImages.shape
# Agregamos el intercepto al conjunto de entrenamiento
XtrainImages = np.column_stack((np.ones(N), XtrainImages))
```

## Construcción del modelo en TensorFlow

En esta sección construiremos el modelo a través de un grafo en TensorFlow. Para esto, construiremos la función de hipótesis, la función de costo y el optimizador (a través de gradiente en descenso de TensorFlow). Finalmente, inicializaremos los parámetros entrenables del modelo. En este caso, se utilizará una función para poder cambiar el grafo de acuerdo al número de parámetros a estimar.

```python
# Crear el grafo
tf.reset_default_graph()
g = tf.Graph()
# Definimos tensores y operaciones en g
with g.as_default():
    # Creando los placeholders
    X = tf.placeholder(tf.float32, shape = (None, K+1), name = "X")
    Ylabels = tf.placeholder(tf.int32, name = "Ylabels")

    # Hiperparámetros del modelo
    lr = tf.placeholder(tf.float32, name = "lr")

    # Coeficientes de regresión de logits, incluyendo intercepto
    W = tf.Variable(tf.truncated_normal(shape = [K+1, 3]), name = "W")

    # Logits de regresión multiple multiclase y pronóstico de clases
    with tf.name_scope("Logits"):
        Logits = tf.matmul(X, W, name = "Logits") 
        YlabelsHat = tf.nn.softmax(Logits, name = "YlabelsHat")

    # Función de costo
    with tf.name_scope("FuncionCosto"):
        cost = tf.reduce_mean(tf.nn.sparse_softmax_cross_entropy_with_logits(labels = Ylabels, logits = Logits), name = "Costo")

    # Gradient Descent Optimizer 
    with tf.name_scope("GradientDes.Optimizer"):
        optimizer = tf.train.GradientDescentOptimizer(lr).minimize(cost) 
        
    # Accuracy del modelo
    with tf.name_scope("Accuracy"):
        correctPredictions = tf.equal(Ylabels, tf.argmax(YlabelsHat, 1, output_type = tf.int32), name = "correctPred")
        accuracy = tf.reduce_mean(tf.cast(correctPredictions, tf.float32), name = "accuracy")
        
    with tf.name_scope("Summaries"):
        # Summary de costo para el TensorBoard
        costSummary = tf.summary.scalar(name='fCosto', tensor = cost)
        # Summary de precisión para el TensorBoard
        accuracySummary = tf.summary.scalar(name='accuracySummary', tensor = accuracy)
        # Agrupar los sumaries
        summaries = tf.summary.merge_all()
    
    # Global Variables Initializer 
    init = tf.global_variables_initializer() 
```

#### Función para obtener un *mini-batch* del conjunto de datos de entrenamiento
La siguiente función nos devuelve el n-ésimo *mini-batch* de la matriz x. La idea es obtener porciones de la matriz de datos de entrenamiento para llevar a cabo el *mini batch gradient descent*.

```python
# Devuelve el batch batchNo de la matriz X, tomado de las filas
def obtenerBatch(x, batchSize, totalBatches, batchNo):
    if batchNo < totalBatches:
        return x[batchSize*batchNo:batchSize*(batchNo+1)]
    else:
        return x[batchSize*totalBatches:]
```

#### Función para entrenamiento de modelo *logit multinomial*
Se utiliza la función `entrenarModeloLineal(x, y, training_epochs, batch_size, learning_rate, imprimirCada)` para llevar a cabo un proceso de minimización con *mini-batch gradient descent*. Los argumentos son: 
* `x` : matriz de imágenes, cada columna es un *feature*.
* `y` : vector de etiquetas, con números enteros de 0 a $m-1$ clases.
* `training_epochs` : número de *epochs* de entrenamiento.
* `batch_size` : tamaño de la muestra de *mini-batch* para el proceso de minimización.
* `imprimirCada` : cada cuántos *batch* imprimir un mensaje de estado.
* `tag` : una etiqueta a agregar para Tensorboard.

```python
def entrenarModeloLineal(x, y, training_epochs, batch_size, learning_rate, imprimirCada, tag):
    # Número total de batches
    totalBatches = (x.shape[0] // batch_size)
    
    # Iniciando la sesión de Tensorflow 
    start = time.time()
    with tf.Session(graph = g) as sess: 
        # Crear el objeto writer
        writer = tf.summary.FileWriter('./logitMultiV2/lr=%0.3f,epochs=%d,batch=%d, <%s>' % (learning_rate, training_epochs, batch_size, tag), sess.graph)
        # Inicializar las variables del grafo 
        sess.run(init) 

        # Iterar para cada epoch 
        for epoch in range(training_epochs): 
            # Iterar para cada uno de los batches
            for batch in range(totalBatches):
                # Obtener el batch
                xbatch = obtenerBatch(x, batch_size, totalBatches, batch)
                ybatch = obtenerBatch(y, batch_size, totalBatches, batch)
                
                # Ejecutar el optimizador de gradiente en descenso y el summary de costo
                _, summ = sess.run([optimizer, summaries], 
                    feed_dict = {X : xbatch, Ylabels : ybatch, lr : learning_rate})

                # Agregar summaries al tablero
                writer.add_summary(summ, epoch*totalBatches + batch + 1)

                # Obtener el costo para imprimir
                if (batch + 1) % imprimirCada == 0: 
                    c = sess.run(cost, feed_dict = {X : xbatch, Ylabels : ybatch})
                    print("Epoch = %d,\tBatch = %d,\tCosto: %0.4f" % (epoch, batch+1, c))
                    
        # Cerrar el writer del summary
        writer.close()
        # Guardar algunos valores para utilizar fuera de la sesión
        weights = sess.run(W)
    
    #Fin de la sesión
    end = time.time()
    print("Tiempo transcurrido: %0.2f segundos. " % (end-start))
    # Devolver los pesos
    return (weights)
```

#### Función para clasificar una imagen
Esta función se utilizará para clasificar una imagen del conjunto de entrenamiento `XtrainImages`. La función devuelve el tipo de prenda como una cadena de texto. Para clasificar, utiliza el argumento `weights`, obtenido del proceso de entrenamiento.

```python
def predecir(x, weights):
    # Clasificar la imagen. No hace falta softmax,
    # basta con ver qué categoría tiene la puntuación más alta
    logits = np.matmul(x, weights)
    y = np.argmax(logits)
    return dicClases[y]
```

#### Función para determinar la etiqueta de un determinado X

```python
def predictLabels(Xtest, Theta):
    u = tf.Graph()
    # Definimos tensores y operaciones en g
    with u.as_default():
        # Creando los placeholders
        X = tf.placeholder(tf.float32, shape = (None, K+1), name = "X")
        #Ylabels = tf.placeholder(tf.int32, name = "Ylabels")

        # Coeficientes de regresión de logits, incluyendo intercepto
        W = tf.placeholder(tf.float32, shape = (K+1, 3), name = "W")

        # Logits de regresión multiple multiclase y pronóstico de clases
        with tf.name_scope("Logits"):
            Logits = tf.matmul(X, W, name = "Logits") 
            YlabelsHat = tf.argmax(tf.nn.softmax(Logits), 1, output_type = tf.int32, name = "YlabelsHat")
        
    # Session
    with tf.Session(graph = u) as sess:
        pred = sess.run(YlabelsHat, feed_dict = {X : Xtest, W : Theta})
    
    return pred
```

#### Función para obtener la precisión de un modelo

```python
def obtenerPrecision(Xtest, YtrueLabels, theta):
    predModelo = predictLabels(Xtest, theta)
    acc = np.mean(np.equal(predModelo, YtrueLabels))
    print("La precisión del modelo es %0.2f%%" % (acc*100))
    return(acc)
```

```python
# Computamos el accuracy del modelo con datos crudos
obtenerPrecision(XtestImages, YtestLabels, Theta);
```

```python
# Computamos el accuracy del modelo con datos normalizados
XtestImagesRenorm = XtestImages.copy()
XtestImagesRenorm[:, 1:] /= 255. 
obtenerPrecision(XtestImagesRenorm, YtestLabels, ThetaRenorm);
```
